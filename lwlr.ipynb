```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sigma = 0.2
X = np.array(np.random.normal(10,5,100))
X = np.sort(X)
func = np.cos((X*np.pi)/3.5)
y = np.array(func+np.random.normal(0,sigma,100))
```

```python
plt.plot(X,y,'ro')
plt.show()
```

```python
X_mat = np.matrix(X).T
m = X_mat.shape[0]
print(m)
y_mat=np.matrix(y).T
```

```python
X_con = np.concatenate([np.ones([m,1]),X_mat],1)
```

```python
def lwlr(X,y,tau):
    y_pred = np.zeros(m)
    for i in range(m):
        y_pred[i]=X[i]*localweight(X[i],X,y,tau)
    return y_pred


def localweight(xi,X,y, tau):
    wt = funct(xi,X,tau)
    w = np.linalg.inv(X.T*(wt*X))*((X.T*wt)*y)
    return w

def funct(xi,X,tau):
    weights = np.mat(np.eye(m))
    for i in range(m):
        diff = xi-X[i]
        weights[i,i]=np.exp(diff*diff.T/(-2.0*tau**2))
    return weights
```

```python
tau = 0.4
y_pred = lwlr(X_con,y_mat,tau)
```

```python
X_Y_pair = np.concatenate([X_mat,np.matrix(y_pred).T],1)
```

```python
plt.plot(X_Y_pair[:,0],X_Y_pair[:,1],'b-')
plt.plot(X,y,'g.')
```

```python
tau = 2
y_pred = lwlr(X_con,y_mat,tau)
X_Y_pair = np.concatenate([X_mat,np.matrix(y_pred).T],1)
plt.plot(X_Y_pair[:,0],X_Y_pair[:,1],'b-')
plt.plot(X,y,'g.')
```
